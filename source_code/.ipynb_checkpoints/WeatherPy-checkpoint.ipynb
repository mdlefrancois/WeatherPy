```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90) #from south pole to north pole
lng_range = (-180, 180) #from west to east
```

## Generate Cities List

```python
#Randomly select at least 500 unique (non-repeat) cities based on latitude and longitude.
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500) #1500
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
#lat_lng[0] is lattitude, lat_lng[1] is longitude
#Latitude values are measured relative to the equator and range from -90째 at the South Pole to +90째 at the North Pole. 
#Longitude values are measured relative to the prime meridian. They range from -180째 when traveling west to 180째 
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    #print(city)
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print(cities)
```

## Perform API Calls

```python
#Perform a weather check on each of the cities using a series of successive API calls.
#Include a print log of each city as it's being processed with the city number and city name.
#Save both a CSV of all data retrieved and png images for each scatter plot.

# Build query Url
base_url = "http://api.openweathermap.org/data/2.5/weather?"
api_keys = api_key
units = "imperial"

query_url = base_url + "appid=" + api_key + "&units=" + units + "&q="
#print(query_url)

# set up lists to hold reponse info
City = []
lat = []
temp = []
Humidity = [] #(%) vs. Latitude
Cloudiness = [] #(%) vs. Latitude
WindSpeed = [] #(mph) vs. Latitude

# Loop through the list of cities and perform a request for data on each
print("---- Looping through calls to OpenWeather API ----")
Count = 0
for city in cities:
    #print(query_url + city)
    response = requests.get(query_url + city).json()
    print("---- looking at city: " + city + "   API count: " + str(Count))
    #pprint(response)
    Count = Count + 1

    try:
        City.append(response['name'])
        lat.append(response['coord']['lat'])
        temp.append(response['main']['temp'])
        Humidity.append(response['main']['humidity'])
        Cloudiness.append(response['clouds']['all'])
        WindSpeed.append(response['wind']['speed'])

        #when the city is not found
        if "{'cod': '404', 'message': 'city not found'}" in response:
            print("City Not Found")
    except:
        continue
            
print("--------------------------------------------------")
print("API Calls Finished")
print("Completed " + str(len(cities)) + " calls")
```

```python
# create a data frame from cities, lat, temp, humidity, cloudiness and windspeed
weather_dict = {
    "city": City,
    "lat": lat,
    "temp": temp,
    "Humidity": Humidity,
    "Cloudiness": Cloudiness,
    "WindSpeed": WindSpeed
}
weather_data = pd.DataFrame(weather_dict)
#weather_data.head(30)

#drop duplicates
weather_data.drop_duplicates(['city'])
weather_data.to_csv('WorldCityLatitude_AtmosphericInfo.csv', index=False, header=True)
```

```python
# Build a scatter plot for temperature versus latitude
#scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths
#=None, verts=None, edgecolors=None, *, data=None, **kwargs)
plt.scatter(weather_data["lat"], weather_data["temp"], color='darkblue',marker="8",alpha=0.75)
plt.axvline(0, color='k')

# Incorporate the other graph properties
plt.title("City Latitude vrs Temperature")
plt.ylabel("Temperature (Fahrenheit)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("TemperatureInWorldCities.png")

# Show plot
plt.show()
```
